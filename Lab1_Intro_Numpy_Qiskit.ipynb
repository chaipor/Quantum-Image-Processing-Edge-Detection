```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
some_state=[0,1]
plot_bloch_multivector(some_state)
```

```python
import numpy as np
a=1+2j
b=1+1j
mag=np.sqrt(abs(a)**2+abs(b)**2)
state2=[a/mag,b/mag]
plot_bloch_multivector(state2)
```

```python
import numpy as np
from qiskit import QuantumCircuit
qc = QuantumCircuit(1)
qc.h(0)


qc.draw('mpl')
```

```python
from qiskit.quantum_info import Statevector
sv = Statevector.from_label('0')
print(sv)
plot_bloch_multivector(sv)
```

```python
new_sv = sv.evolve(qc)
print(new_sv)
plot_bloch_multivector(new_sv)
```
